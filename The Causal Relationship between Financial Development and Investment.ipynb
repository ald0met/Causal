# 그랜저 인과관계 분석

## 개요
그랜저 인과관계(Granger causality) 검정은 두 시계열 변수 간의 인과관계를 평가하는 통계적 방법입니다. 그러나 실제로는 인과관계보다는 상관관계에 기반한 분석을 수행합니다.

### 그랜저 인과관계의 개념
1. **정의**:
   - 변수 A가 변수 B의 미래 값을 예측하는 데 유용한 정보를 제공하는 경우, A가 B를 "그랜저 원인"한다고 말합니다.

2. **상관관계와 인과관계**:
   - 그랜저 인과관계 검정은 두 변수 간의 시간적 선후 관계를 분석하지만, 인과관계를 직접적으로 증명하지 않습니다.

### 한계
1. **인과관계의 명확한 증명 부족**:
   - 그랜저 인과관계 검정은 인과관계를 명확히 증명하지 않으며, 상관관계가 인과관계로 해석될 수 있습니다.

2. **시간 지연**:
   - 시간 지연을 고려하지만, 잘못된 해석이 발생할 수 있습니다.

3. **모델의 적합성**:
   - 특정 모델에 기반하여 수행되므로, 모델의 적합성이 결과에 영향을 미칠 수 있습니다.

# 그랜저 인과관계와 상관관계 분석

## 개요
그랜저 인과관계(Granger causality) 검정은 두 시계열 변수 간의 인과관계를 평가하는 통계적 방법입니다. 그러나 실제로는 인과관계보다는 상관관계에 기반한 분석을 수행합니다. 이 분석에서는 두 시계열 데이터 간의 상관관계를 시각화하고, 그랜저 인과관계 검정을 통해 두 변수 간의 관계를 비교합니다.

## 그랜저 인과관계가 결국 상관관계 분석에 불과하다는 이유

### 1. 상관관계의 존재
- **상관관계 행렬**: 두 변수 간의 상관관계를 계산하여 출력합니다. 상관관계 행렬은 두 변수 간의 선형 관계의 강도를 나타내며, 값이 1에 가까울수록 강한 양의 상관관계를, -1에 가까울수록 강한 음의 상관관계를 의미합니다. 만약 A와 B 간의 상관관계가 높다면, 이는 두 변수 간에 어떤 관계가 있음을 나타냅니다.

### 2. 그랜저 인과관계 검정
- **그랜저 인과관계 검정 결과**: 그랜저 인과관계 검정은 A의 과거 값이 B의 현재 값을 예측하는 데 유용한지를 평가합니다. 그러나 이 검정은 인과관계를 명확히 증명하는 것이 아니라, 단지 A의 과거 값이 B에 대한 예측력을 제공하는지를 확인합니다. 만약 A와 B 간의 상관관계가 높다면, 그랜저 인과관계 검정에서 A가 B를 그랜저 원인한다고 결론지을 가능성이 높습니다.

### 3. 상관관계와 인과관계의 차이
- **인과관계의 해석**: 그랜저 인과관계 검정의 결과가 A가 B를 그랜저 원인한다고 나타나더라도, 이는 반드시 A가 B에 영향을 미친다는 것을 의미하지 않습니다. 두 변수 간의 상관관계가 높은 경우, B가 A에 영향을 미치거나, 두 변수 모두 다른 외부 요인에 의해 영향을 받을 수 있습니다. 즉, 상관관계가 인과관계로 해석될 수 있는 가능성이 존재합니다.

## 결론
결국, 그랜저 인과관계 검정은 두 변수 간의 상관관계를 기반으로 하며, 인과관계의 방향성을 추론하는 데 사용됩니다. 그러나 상관관계가 인과관계를 의미하지 않기 때문에, 그랜저 인과관계 검정의 결과를 해석할 때는 주의가 필요합니다.

# 인과성을 강건하게 확인하는 방법

사후 확률 계산 이후에 인과성을 더 강건하게 확인하기 위해 사용할 수 있는 여러 방법이 있습니다. 이러한 방법들은 인과 관계의 신뢰성을 높이고, 결과에 대한 원인 요인의 영향을 보다 명확히 평가하는 데 도움을 줍니다.

## 1. 감도 분석 (Sensitivity Analysis) (강화할 수 있는 방법)
- **정의**: 모델의 결과가 입력 변수의 변화에 얼마나 민감한지를 평가합니다.
- **방법**: 각 요인(a, b, c)의 사전 확률이나 우도를 변화시켜 결과에 미치는 영향을 분석합니다. 이를 통해 특정 요인이 결과에 미치는 영향이 얼마나 강한지를 평가할 수 있습니다.

## 2. 대체 모델링 (Alternative Modeling) (참고)
- **정의**: 다양한 통계적 모델을 사용하여 인과 관계를 평가합니다.
- **방법**: Bayesian 네트워크 외에도 회귀 분석, 구조적 방정식 모델링(SEM), 기계 학습 모델 등을 사용하여 인과 관계를 검증합니다. 여러 모델에서 일관된 결과가 나타나면 인과성이 더 강건하다고 판단할 수 있습니다.

## 3. 무작위 대조 실험 (Randomized Controlled Trials, RCT) (전제)
- **정의**: 무작위로 참가자를 처치 그룹과 대조 그룹에 배정하여 인과 관계를 평가하는 실험적 방법입니다.
- **방법**: RCT를 통해 직접적인 인과 관계를 확인할 수 있습니다. RCT는 외부 요인의 영향을 최소화하여 인과성을 명확히 할 수 있는 강력한 방법입니다.

## 4. 도구 변수 (Instrumental Variables) (강화할 수 있는 방법)
- **정의**: 인과 관계를 추정하기 위해 사용되는 외부 변수입니다.
- **방법**: 도구 변수를 사용하여 원인과 결과 간의 인과 관계를 추정합니다. 도구 변수는 원인에 영향을 미치지만 결과에는 직접적인 영향을 미치지 않아야 합니다.

## 5. 시간적 순서 (Temporal Ordering) (Granger)
- **정의**: 원인과 결과 간의 시간적 관계를 평가합니다.
- **방법**: 원인이 결과보다 먼저 발생해야 인과 관계가 성립합니다. 시간적 순서를 명확히 하여 인과성을 강화할 수 있습니다.

## 6. 조건부 독립성 검토 (전제)
- **정의**: 특정 조건 하에서 두 변수 간의 독립성을 평가합니다.
- **방법**: Bayesian 네트워크에서 조건부 독립성을 검토하여 원인과 결과 간의 관계를 더 명확히 할 수 있습니다. 특정 요인이 주어졌을 때 다른 요인과 결과가 독립적인지를 평가합니다.

## 7. 다양한 데이터 소스 통합 (3과의 차이 ?)
- **정의**: 여러 출처의 데이터를 통합하여 인과 관계를 평가합니다.
- **방법**: 관찰적 데이터와 실험적 데이터를 결합하여 인과 관계를 더 강건하게 검증할 수 있습니다. 다양한 데이터 소스에서 일관된 결과가 나타나면 인과성이 더 신뢰할 수 있습니다.

## 8. 모델 검증 (Model Validation) (참고)
- **정의**: 모델의 예측력을 평가합니다.
- **방법**: 교차 검증, 후속 연구, 외부 데이터셋을 사용하여 모델의 예측력을 검증합니다. 모델이 새로운 데이터에서도 일관된 결과를 보인다면 인과성이 더 강건하다고 판단할 수 있습니다.
